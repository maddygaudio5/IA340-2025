```python
df = df[['Incident ID','Victims','Address','State']] 
df.head()
```

Keep specific columns in this case Incident ID, Victims, and Address

```python
sum_victims_by_state = df.groupby('State')['Victims'].sum()
sum_victims_by_state.plot.bar()
```

Maryland holds the greatest amount of victims in comparison to other states. Outscored the other states by more than 40,000 victims.
